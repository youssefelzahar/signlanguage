```python
import cv2
import os

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = plt.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
            plt.imshow(img)
            plt.show()
    return images
```

```python
#load_images_from_folder("dataSet/trainingData/A")
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
train_set=train_datagen.flow_from_directory("dataSet/trainingData",
                                            target_size=(128,128),
                                            batch_size=10,
                                            color_mode="grayscale",
                                            class_mode="categorical",
                                            shuffle=True,
                                            seed=42)
                                            
                            
```

```python
test_set=test_datagen.flow_from_directory("dataSet/testingData",
                                            target_size=(128,128),
                                            batch_size=10,
                                            color_mode="grayscale",
                                            class_mode="categorical",
                                            shuffle=True,
                                            seed=42)
                                            
```

```python
for _ in range(5):
   
    img, label = train_set.next()
    print(img.shape) 
    plt.figure(figsize=(2,2))#  (1,256,256,3)
    plt.imshow(img[0])
    plt.show()
```

```python
for _ in range(5):
    img, label = test_set.next()
    print(img.shape) 
    plt.figure(figsize=(2,2))#  (1,256,256,3)
    plt.imshow(img[0])
    plt.show()
```

```python
def embding():
    inp=Input(shape=(None, 128, 128, 3), dtype=")
    
    conv1=Conv2D(64,kernel_size=3,padding="same",activation="relu")(inp)
    pooling1=MaxPool2D(2,2,padding="valid")(conv1)
    
    conv2=Conv2D(64,kernel_size=3,padding="same",activation="relu")(pooling1)
    pooling2=MaxPool2D(2,2,padding="valid")(conv2)
    
    #conv3=Conv2D(64,kernel_size=3,padding="same",activation="relu")(pooling2)
    
    #pooling3=MaxPool2D(2,2,padding="valid")(conv3)
    
    f1=Flatten()(pooling2)
    
    d1=Dense(128,activation="relu")(f1)
    drop1=Dropout(0.40)(d1)
    d2=Dense(96,activation="relu")(drop1)
    drop2=Dropout(0.40)(d2)
    d3=Dense(64,activation="relu")(drop2)
    d4=Dense(27,activation="softmax")(d3)
   
    return Model(inputs=[inp],outputs=[d4],name="embedding")


    
```

```python
model=embding()
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', 
                   loss = 'categorical_crossentropy', 
                   metrics = ['accuracy'])
```

```python
history=model.fit(train_set,
                  epochs = 10,
                  validation_data = test_set)
```

```python
plt.figure(figsize=[8,6])
plt.plot(history.history['loss'],'r',linewidth=3.0)
plt.plot(history.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
```

```python
plt.figure(figsize=[8,6])
plt.plot(history.history['accuracy'],'r',linewidth=3.0)
plt.plot(history.history['val_accuracy'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
```
